## 로봇 청소기

```python
# 벽또는 빈칸이다. 
# 청소기는 바라보는 방향이 존재 동1/서3/남2/북0 중 하나
# 지도의 각칸은 r,c로 나타내고 r은 북쪽으로 부터 떨어진 칸 개수 C는 서쪽으로 부터 떨어진 칸의 개수
# 시계방향으로 탐색한다.
'''
1. 현재 위치 청소 
2. 현재 위치에서 현재 방향을 기준으로 왼쪽 방향 부터 탐색 
2-1. 왼쪽 청소공간 O -> 그방향으로 회전 한칸 전진 1번 진행
2-2. 왼쪽 청소공간 X -> 그방향으로 회전하고 2번으로 돌아간다 
2-3. 네방향 모두 청소O/벽 -> 바라보는 방향 유지/한칸 후진 2번 돌아감
2-4. 네방향 모두 청소O/벽  뒤쪽 방향 벽이라 후진도X 멈춘다. 
'''

from collections import deque 

n,m=map(int,input().split())
matrix=[]
x,y,d=map(int,input().split())

for _ in range(n):
  matrix.append(list(map(int,input().split())))

visited=[[False]*(m) for _ in range(n)]
visited[x][y]=True 

# 북0 동1 남2 서3
dx=[-1,0,1,0]
dy=[0,1,0,-1]
cnt=0

while True:
  flag=0
  for _ in range(4):
    nx,ny=x+dx[(d+3)%4],y+dy[(d+3)%4]
    d=(d+3)%4
    if 0<=nx<n and 0<=ny<m and matrix[nx][ny]==0 and not visited[nx][ny]:
      x,y=nx,ny
      cnt+=1
      flag=1
      break
  if not flag:
    nx,ny=x-dx[(d+3)%4],y-dy[(d+3)%4]
    if 0<=nx<n and 0<=ny<m and matrix[nx][ny]==0 and not visited[nx][ny]:
      x,y=nx,ny
      continue
    else:
      break 
    
print(cnt)

  

  
```

```python
from collections import deque
from itertools import combinations
import copy

'''

1. dq에 값을 저장해놓는다. 순열을 이용하여 순서쌍을 넣어놓는다.
1~6까지 들어있는값들을

2. dq에 저장되어있는것들을 한번씩 하여 수행한다. 

3. 이때 matrix를 새로 만든다음에 cctv가 역할을 수행하였을때 남는 0의 개수를 센다
4. 0의 개수를 세서 반환을 한뒤에 앞으로 이 값을 비교하여 최소가 될때를 구한다. 

'''

n, m = map(int, input().split())
matrix,graph = [],[]
arr = []

for i in range(n):
    temp = list(map(int, input().split()))
    for j in range(len(temp)):
        if temp[j] in [1,2,3,4,5]:
            arr.append((i, j))
    graph.append(temp)


dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

# 1번이라고 가정하자

def foward(x, y, dir):
    dq = deque()
    dq.append((x, y))
    cnt = 0
    while dq:
        x, y = dq.popleft()
        nx, ny = x + dx[dir], y + dy[dir]
        if 0>nx or n<=nx or 0>ny or m<=ny:
            continue
        if matrix[nx][ny] == 0 and 0 <= nx < n and 0 <= ny < m:
            dq.append((nx, ny))
            cnt += 1
    return cnt

def express(x, y, dir):
    dq = deque()
    dq.append((x, y))
    cnt = 0
    while dq:
        x, y = dq.popleft()
        nx, ny = x + dx[dir], y + dy[dir]
        if 0>nx or n<=nx or 0>ny or m<=ny:
            continue
        if matrix[nx][ny] == 0 and 0 <= nx < n and 0 <= ny < m:
            matrix[nx][ny]="#"
            dq.append((nx, ny))

def two(x,y,dir):
    cnt=0
    cnt+=foward(x,y,dir)
    dir=(dir + 2) % 4
    cnt+=foward(x,y,dir)
    return cnt

def three(x,y,dir):
    cnt=0
    cnt+=foward(x,y,dir)
    dir=(dir + 1) % 4
    cnt+=foward(x,y,dir)

def four(x,y,dir):
    cnt=0
    cnt+=foward(x,y,dir)
    dir=(dir + 1) % 4
    cnt+=foward(x,y,dir)
    dir=(dir + 1) % 4
    cnt+=foward(x,y,dir)

n = len(arr)
final_res=int(1e9)
for c in combinations(arr, n):
    matrix=copy.deepcopy(graph)
    dq=deque(c)
    while dq:
        i,j=dq.popleft()
        if matrix[i][j] not in [1,2,3,4,5]:
            continue
        x,y,max_dir=0,0,0
        cnt,res=0,0
        no=matrix[i][j]
        if no==5:
            for k in range(4):
                express(i,j,k)
            continue
        for k in range(4):
            if no==1:
                cnt = foward(i, j, k)
                if res < cnt:
                    res = cnt
                    x, y, max_dir = i, j, k
            elif no==2:
                cnt = two(i, j, k)
                if res < cnt:
                    res = cnt
                    x, y, max_dir = i, j, k
            elif no==3:
                cnt = three(i, j, k)
                if res < cnt:
                    res = cnt
                    x, y, max_dir = i, j, k
        if no==1:
            express(x,y,max_dir)
        elif no==2:
            express(x,y,max_dir)
            max_dir = (max_dir + 2) % 4
            express(x,y,max_dir)
        elif no==3:
            express(x,y,max_dir)
            max_dir = (max_dir + 1) % 4
            express(x,y,max_dir)
        elif no==4:
            express(x,y,max_dir)
            max_dir = (max_dir + 1) % 4
            express(x,y,max_dir)
            max_dir = (max_dir + 1) % 4
            express(x,y,max_dir)

        for s in matrix:
            print(s)

    res=0
    for i in range(n):
        for j in  range(m):
            if matrix[i][j]==0:
                res+=1
    final_res=min(final_res,res)

print(final_res)
```

```python
n,m,t=map(int,input().split())
# https://kyun2da.github.io/2021/04/20/brownsmog/
matrix=[]

air_cond=[]
for i in range(n):
  temp=list(map(int,input().split()))
  matrix.append(temp)

for i in range(n):
  if matrix[i][0]==-1:
    up=i
    break

dx=[-1,1,0,0]
dy=[0,0,-1,1]

down=up+1
def disperse():
  dmatrix=[[0]*(m) for _ in range(n)]
  for x in range(n):
    for y in range(m):
      if matrix[x][y]!=0 and matrix[x][y]!=-1:
        cnt=0
        init=matrix[x][y]//5
        for k in range(4):
          nx,ny=x+dx[k],y+dy[k]
          if 0>nx or n<=nx or 0>ny or m<=ny or matrix[nx][ny]==-1:
            continue
          if matrix[nx][ny]!=-1:
            dmatrix[nx][ny]+=init
            cnt+=1
        matrix[x][y]-=init*cnt
        
  for i in range(n):
    for j in range(m):
      matrix[i][j]+=dmatrix[i][j]


def air_up():
  global up
  dx = [0, -1, 0, 1]
  dy = [1, 0, -1, 0]
  direct,before=0,0
  x,y=up,1
  while True:
    nx,ny=x+dx[direct],y+dy[direct]
    if x==up and y==0:
      break
    if nx<0 or nx>=n or ny<0 or ny>=m:
      direct+=1
      continue
    matrix[x][y],before=before,matrix[x][y]
    x,y=nx,ny

def air_down():
  dx = [0, 1, 0, -1]
  dy = [1, 0, -1, 0]
  direct = 0
  before = 0
  x, y = down, 1
  while True:
    nx = x + dx[direct]
    ny = y + dy[direct]
    if x == down and y == 0:
      break
    if nx < 0 or nx >= n or ny < 0 or ny >=m:
      direct += 1
      continue
    matrix[x][y], before = before, matrix[x][y]
    x,y = nx,ny

for _ in range(t):
  disperse()
  air_up()
  air_down()

res=0
for i in range(n):
  for j in range(m):
    if matrix[i][j]>0:
      res+=matrix[i][j]
print(res)
```
